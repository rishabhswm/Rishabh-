Part1: Numpy Fundamentals

```
import numpy as np
# A 1D array of integers from 10 to 50 (step of 5)
array1 = np.arange(10, 51, 5)
print("1D array:", array1)

# A 3x4 matrix of zeros
zeros_matrix = np.zeros((3, 4))
print("3x4 matrix of zeros:\n", zeros_matrix)

# A 2x3 matrix of random integers between 1 and 20
random_matrix = np.random.randint(1, 21, size=(2, 3))
print("2x3 random matrix:\n", random_matrix)

# An Identity matrix of size 5x5
identity_matrix = np.eye(5)
print("5x5 Identity matrix:\n", identity_matrix)

# For the 2x3 random matrix:
print("Shape:", random_matrix.shape)
print("Size:", random_matrix.size)
print("Data type:", random_matrix.dtype)
print("Sum of all elements:", np.sum(random_matrix))
print("Max value:", np.max(random_matrix))
print("Index of max value:", np.unravel_index(np.argmax(random_matrix), random_matrix.shape))

# Create 6x6 matrix with values 1 to 36
matrix6x6 = np.arange(1, 37).reshape(6, 6)
print("6x6 matrix:\n", matrix6x6)

# Extract and print:
print("Third row:", matrix6x6[2])
print("Last two columns:\n", matrix6x6[:, -2:])
print("All elements greater than 20:\n", matrix6x6[matrix6x6 > 20])
print("Every second element in the diagonal:\n", matrix6x6.diagonal()[::2])

# Replace all values in first and last rows with -1
matrix6x6[0, :] = -1
matrix6x6[-1, :] = -1
print("Modified matrix:\n", matrix6x6)
a = np.array([1, 2, 3, 4, 5])
b = np.array([2, 4, 6, 8, 10])

# Element-wise operations
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)

# Dot product
print("Dot product:", np.dot(a, b))

# Square root of array a
print("Square root of a:", np.sqrt(a))

# Natural logarithm of array b
print("Log of b:", np.log(b))
# Generate a 4x5 array from a normal distribution
stat_array = np.random.normal(loc=50, scale=15, size=(4, 5))
print("4x5 Random Array:\n", stat_array)

# Statistics
print("Mean of entire array:", np.mean(stat_array))
print("Median:", np.median(stat_array))
print("Standard deviation:", np.std(stat_array))

# Mean of each row
print("Mean of each row:", np.mean(stat_array, axis=1))

# Max value in each column
print("Max in each column:", np.max(stat_array, axis=0))

# Correlation coefficient between first and last rows
corr_coeff = np.corrcoef(stat_array[0], stat_array[-1])[0, 1]
print("Correlation coefficient between first and last row:", corr_coeff)
```

Part 2: Advanced Numpy Application

```
import numpy as np

# 1. Create a 1D array with 24 elements
arr = np.arange(1, 25)
print("1D Array:\n", arr)

# 2. Reshape
print("2x12:\n", arr.reshape(2, 12))
print("3x8:\n", arr.reshape(3, 8))
print("4x6:\n", arr.reshape(4, 6))
print("2x3x4:\n", arr.reshape(2, 3, 4))

# 3. Broadcasting
a = np.array([[10], [20], [30]])  # 3x1
b = np.array([1, 2, 3, 4])        # 1x4

# 4. Add
broadcast_result = a + b
print("Broadcasting Result:\n", broadcast_result)

# 5. Explanation:
# Shape of a: (3, 1), b: (4,) → broadcasted to (3, 4)
# Each row in 'a' is added to the array 'b'
arr = np.random.randint(1, 101, (5, 5))
print("Original 5x5 Array:\n", arr)

# Elements > 50
print("Elements > 50:\n", arr[arr > 50])

# Replace all even numbers with negative values
arr[arr % 2 == 0] *= -1
print("After replacing even numbers with negatives:\n", arr)

# Count how many divisible by 3
count_div3 = np.sum(arr % 3 == 0)
print("Count divisible by 3:", count_div3)

# np.where()
new_arr = np.where(arr > 75, 1, np.where(arr < 25, -1, 0))
print("np.where array:\n", new_arr)
# Create matrices
A = np.random.rand(3, 3)
B = np.random.rand(3, 3)
print("Matrix A:\n", A)
print("Matrix B:\n", B)

# Matrix multiplication
product = np.dot(A, B)
print("Dot Product:\n", product)

# Determinant and Inverse of A
det_A = np.linalg.det(A)
inv_A = np.linalg.inv(A)
print("Determinant of A:", det_A)
print("Inverse of A:\n", inv_A)

# Equations:
# 2x + 3y = 7
# 4x - y = 1
coeff = np.array([[2, 3], [4, -1]])
const = np.array([7, 1])
solution = np.linalg.solve(coeff, const)
print("Solution to linear system:\n", solution)
```

Part 3: Matplotlib Visualization

```
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
# Line Plot
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin(x)", linestyle='--', color='blue')
plt.plot(x, y2, label="cos(x)", linestyle='-', color='orange')
plt.title("Trigonometric Functions")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()

# Scatter Plot
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, c=y, s=100*y, alpha=0.7)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Random Scatter Plot")
plt.show()
# Histogram
data = np.random.normal(loc=0, scale=1, size=1000)
sns.histplot(data, bins=30, kde=True)
plt.title("Histogram with Density Curve")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Box Plot for 4 groups
group_data = [np.random.normal(loc=mu, size=100) for mu in [20, 50, 75, 100]]
plt.boxplot(group_data, patch_artist=True)
plt.xticks([1, 2, 3, 4], ["Group A", "Group B", "Group C", "Group D"])
plt.title("Box Plot of 4 Groups")
plt.show()
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Bar Chart (Top-left)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [200, 250, 300, 220, 310]
axs[0, 0].bar(months, sales, color='skyblue')
axs[0, 0].set_title("Monthly Sales")

# Pie Chart (Top-right)
categories = ['A', 'B', 'C', 'D', 'E']
shares = [20, 25, 15, 30, 10]
axs[0, 1].pie(shares, labels=categories, autopct='%1.1f%%')
axs[0, 1].set_title("Market Share")

# Heatmap (Bottom-left)
data = np.random.rand(5, 5)
sns.heatmap(data, ax=axs[1, 0], annot=True, cmap='coolwarm')
axs[1, 0].set_title("5x5 Correlation Matrix")

# 3D Surface Plot (Bottom-right)
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

ax = fig.add_subplot(224, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title("3D Surface Plot: z = sin(x) * cos(y)")

plt.tight_layout()
plt.show()
```

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define df
np.random.seed(0)
dates = pd.date_range(start="2024-01-01", end="2024-12-31")
temps = np.random.normal(loc=25, scale=7, size=len(dates))  # 365 values


df = pd.DataFrame({'Date': dates, 'Temperature': temps})
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
monthly_avg = df.groupby('Month')['Temperature'].mean()
print("Monthly Averages:\n", monthly_avg)
hottest = df[df['Temperature'] == df['Temperature'].max()]
coldest = df[df['Temperature'] == df['Temperature'].min()]

print("\nHottest Day:\n", hottest)
print("\nColdest Day:\n", coldest)
# Using a simple moving average to identify trend
df['7_day_avg'] = df['Temperature'].rolling(window=7).mean()


plt.figure(figsize=(12, 4))
plt.plot(df['Date'], df['Temperature'], label="Daily Temp", alpha=0.5)
plt.title("Daily Temperature Over a Year")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.grid(True)
plt.legend()
plt.show()

#
plt.figure(figsize=(10, 5))
sns.boxplot(x='Month', y='Temperature', data=df)
plt.title("Monthly Temperature Boxplot")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.show()
#
plt.figure(figsize=(8, 5))
plt.hist(df['Temperature'], bins=30, color='skyblue', edgecolor='black')
plt.title("Temperature Distribution")
plt.xlabel("Temperature (°C)")
plt.ylabel("Frequency")
plt.show()
#
plt.figure(figsize=(12, 4))
plt.plot(df['Date'], df['Temperature'], label="Daily Temp", alpha=0.3)
plt.plot(df['Date'], df['7_day_avg'], label="7-Day Avg", color='red', linewidth=2)
plt.title("Moving Average Temperature Trend")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.show()
```
